```
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle
import sys

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

DATADIR = os.path.join(os.getcwd(), "drive", "My Drive", "TU20 Cup", "Old", "AI")
CATEGORIES = ["Desert", "At Risk", "Safe", "Urban"]
```

```
training_data = []
test_data = []

def create_data():
  for category in CATEGORIES:
    path = os.path.join(DATADIR, "Train Images", category)
    class_num = CATEGORIES.index(category)
    count = 0
    print("Train ("+category+"): ", end="")
    for img in os.listdir(path):
      count += 1
      img_array = cv2.imread(os.path.join(path, img))
      #Convert image from BGR (cv2 format) to RGB (plt format)
      img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      #sketchy_array = np.rot90(img_array, k=2, axes=(0,1))
      #plt.imshow(img_array)
      #plt.show()
      #plt.imshow(sketchy_array)
      #plt.show()
      print(str(count)+" ", end="")
      for i in range(0, 4):
        new_array = np.rot90(img_array, k=i, axes=(0,1))
        training_data.append([new_array, class_num])
    path = os.path.join(DATADIR, "Test Images", category)
    count = 0
    print()
    print("Test ("+category+"): ", end="")
    for img in os.listdir(path):
      count += 1
      img_array = cv2.imread(os.path.join(path, img))
      img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      print(str(count)+" ", end="")
      for i in range(0, 1):
        new_array = np.rot90(img_array, k=i, axes=(0,1))
        test_data.append([new_array, class_num])
    print()
  random.shuffle(training_data)
  random.shuffle(test_data)

create_data()
```

```
train_X = []
train_y = []
test_X = []
test_y = []

for features, label in training_data:
  train_X.append(features)
  train_y.append(label)
for i in range(0, len(train_X)):
  train_X[i] = np.array(train_X[i]).reshape(256, 256, 3)

for features, label in test_data:
  test_X.append(features)
  test_y.append(label)
for i in range(0, len(test_X)):
  test_X[i] = np.array(test_X[i]).reshape(256, 256, 3)
train_X = np.array(train_X).reshape(-1, 256, 256, 3)
train_y = np.array(train_y).reshape(-1, 1)
test_X = np.array(test_X).reshape(-1, 256, 256, 3)
test_y = np.array(test_y).reshape(-1, 1)
```

```
np.save(os.path.join(DATADIR, "Train Images", "train_X.npy"), train_X)
np.save(os.path.join(DATADIR, "Train Images", "train_y.npy"), train_y)
np.save(os.path.join(DATADIR, "Test Images", "test_X.npy"), test_X)
np.save(os.path.join(DATADIR, "Test Images", "test_y.npy"), test_y)
print(len(train_y))
print(len(test_y))
#plt.imshow(train_X[0])
#plt.show()
```
